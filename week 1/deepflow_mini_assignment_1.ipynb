```python
df_exploded = data_copy.copy()
df_exploded['cuisines'] = df_exploded['cuisines'].str.split(', ')
df_exploded = df_exploded.explode('cuisines')
df_exploded
```

```python
cleaned_data_path = "/content/drive/My Drive/Deepflow/week1/zomato_cleaned.xlsx"
df_exploded.to_excel(cleaned_data_path, index=False)
```

```python
df_exploded['cuisines'].value_counts()
```

```python
#Identify top cuisines based on ratings and popularity (votes).
cuisine_grouped = df_exploded.groupby('cuisines').agg({
    'rating': 'mean',
    'votes': 'mean',
    'approx_cost_for_two_people': 'mean'
})
overall_votes_avg = cuisine_grouped['votes'].mean()
filtered = cuisine_grouped[cuisine_grouped['votes'] > overall_votes_avg]
sorted = filtered.sort_values(by='rating', ascending=False)
sorted.round(2).head(10)
```

```python
#Highlight best-rated restaurants
best_rated_restro = data_new.copy()
best_rated_restro = best_rated_restro[best_rated_restro['votes'] > overall_votes_avg]

best_rated_restro = best_rated_restro.sort_values(by=['rating','votes'],ascending=[False,False])
# and link them to their cuisine types.
best_rated_restro = best_rated_restro[['name','rating','votes','cuisines','approx_cost_for_two_people']]
best_rated_restro = best_rated_restro.drop_duplicates('name')
best_rated_restro.reset_index(drop=True, inplace=True)
#display
best_rated_restro.head(10)
```
